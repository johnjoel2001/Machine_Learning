```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn .datasets import make_classification
```

```python
# Making a custom datset
X,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=12)
```

```python
df1=pd.DataFrame(X,columns=['feature1','feature2'])
df2=pd.DataFrame(y,columns=['target'])
```

```python
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df.target.value_counts()
```

## We can clearly see a data imbalance

```python
plt.scatter(final_df.feature1,final_df.feature2,c=final_df.target)
```

## Transfrom the dataset using SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE()
X_smote,y_smote=smote.fit_resample(final_df[['feature1','feature2']],final_df.target)
```

```python
X_smote.shape,y_smote.shape, X.shape,y.shape # Before and After SMOTE
```

```python
len(y[y==0]),len(y[y==1])
```

```python
df1_smote=pd.DataFrame(X_smote,columns=['feature1','feature2'])
df2_smote=pd.DataFrame(y_smote,columns=['target'])
smote_df=pd.concat([df1_smote,df2_smote],axis=1)
```

```python
smote_df.shape
```

```python
plt.scatter(smote_df.feature1,smote_df.feature2,c=smote_df.target)
```
